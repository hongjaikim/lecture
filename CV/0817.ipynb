```python
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print('camera err')
    sys.exit()

f_flag=False
i_flag=False

while True:
    ret,frame = cap.read()
    
    if not ret:
        print('frame err')
        break
    if f_flag:
        frame = cv2.flip(frame,1)
    if i_flag:
        frame=cv2.Canny(frame,50,150)


    cv2.imshow('image',frame)

    
    key = cv2.waitKey(20)
    if key==27:
        break
    elif key==ord('f'):
        f_flag=not f_flag
    elif key==ord('i'):
        i_flag=not i_flag

cap.release()
cv2.destroyAllWindows()
```

```python
def call_trackbar(pos):
    img[:]=pos
    cv2.imshow('image',img)

img = np.ones((400,600),np.uint8)

cv2.imshow('image',img)

# createTrackbar(trackbarName, windowName, value, count, onChange) -> None
cv2.createTrackbar('level','image',img[0][0],255,call_trackbar)

cv2.waitKey()
cv2.destroyAllWindows()
```
