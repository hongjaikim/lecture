```python
import numpy as np
import cv2
import sys
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./fig/puppy.bmp',cv2.IMREAD_COLOR)

if img is None:
    print('read err')
    sys.exit()
    
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(20,20))
plt.subplot(131),plt.imshow(img),plt.axis('off'),plt.title('img_bgr')
plt.subplot(132),plt.imshow(img_rgb),plt.axis('off'),plt.title('img_rgb')
plt.subplot(133),plt.imshow(img_gray,cmap='gray'),plt.axis('off'),plt.title('img_gray')
plt.show()
```

```python
img = cv2.imread('./fig/puppy.bmp',cv2.IMREAD_COLOR)

img_resize = cv2.resize(img,dsize=(1000,1000))

cv2.imwrite('puppy_rs.jpg',img_resize)

if img is None:
    print('read err')
    sys.exit()
    
cv2.namedWindow('puppy',cv2.WINDOW_NORMAL)

cv2.imshow('puppy',img)
cv2.imshow('puppy_rs',img_resize)


cv2.waitKey()
cv2.destroyAllWindows()
```
