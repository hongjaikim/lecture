```python
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./fig/puppy.bmp',cv2.IMREAD_COLOR)
img_gray = cv2.imread('./fig/puppy.bmp',cv2.IMREAD_GRAYSCALE)

# cvtColor(src, code[, dst[, dstCn]]) -> dst
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

if img is None:
    print('read img err')
    sys.exit()

plt.figure(figsize=(20,20))
plt.subplot(131), plt.imshow(img),plt.axis('off'),plt.title('img_bgr')
plt.subplot(132), plt.imshow(img_rgb),plt.axis('off'),plt.title('img')
plt.subplot(133), plt.imshow(img_gray,cmap='gray'),plt.axis('off'),plt.title('img_gray')

plt.show()

cv2.imshow('image',img)

while True:
    k = cv2.waitKey()
    if k==27 or k==ord('q'):
        break

cv2.destroyAllWindows()
```
