선형회귀

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x=np.array([2,4,6,8])
y=np.array([81,93,91,97])
```

```python
plt.scatter(x,y)
plt.show()
```

```python
a=0
b=0
```

```python
lr=0.03
```

```python
epochs=2001
```

```python
n=len(x)
```

```python
for i in range(epochs):
  y_pred = a*x+b
  error=y-y_pred

  a_diff = (2/n)*sum(-x*error)
  b_diff = (2/n)*sum(-(error))

  a=a-lr*a_diff
  b=b-lr*b_diff

  if i%100==0:
    print('epochs={}, 기울기={:.04f}, 절편={:.04f}'.format(i,a,b))
  y_pred = a*x+b

plt.scatter(x,y)
plt.plot(x,y_pred,'r')
plt.show()
```

다중선형회귀

```python
x1=np.array([2,4,6,8])
x2=np.array([0,4,2,3])
y=np.array([81,93,91,97])
```

```python
fig = plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter3D(x1,x2,y)
plt.show()
```

```python
a1=0
a2=0
b=0
lr=0.01
epochs=2001
n=len(x1)

for i in range(epochs):
  y_pred=a1*x1+a2*x2+b
  error = y-y_pred

  a1_diff = (2/n)*sum(-x1*error)
  a2_diff = (2/n)*sum(-x2*error)
  b_diff = (2/n)*sum(-(error))

  a1=a1-lr*a1_diff
  a2=a2-lr*a2_diff
  b=b-lr*b_diff

  if i%100==0:
    print('epochs={}, 기울기17={:.04f},기울기2={:.04f}, 절편={:.04f}'.format(i,a1,a2,b))
print(y,y_pred,sep='\n')
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x=np.array([2,4,6,8])
y=np.array([81,93,91,97])

model=Sequential()

model.add(Dense(1,input_dim=1,activation='linear'))

model.compile(optimizer='sgd',loss='mse')

model.fit(x,y,epochs=2000)

plt.scatter(x,y)
plt.plot(x,model.predict(x),'r')
plt.show()

hour=7
prediction=model.predict([hour])
print('%.f, %.02f'%(hour,prediction))
```

다중회귀

```python
x=np.array([[2,0],[4,4],[6,2],[8,3]])
y=np.array([81,93,91,97])
```

```python
model = Sequential()
```

```python
model.add(Dense(1,input_dim=2,activation='linear'))
model.compile(optimizer='sgd',loss='mse')
model.fit(x,y,epochs=2000)

```

```python
hour = 7
private_class=4
prediction = model.predict([[hour,private_class]])

print('%.f,%.f,%.02f'%(hour,private_class,prediction))
```
